```python
# Import our dependencies
import pandas as pd
```

## US Company Information 

```python
# Read in our US Companies data
us_companies_df = pd.read_csv("US_companies.csv")

us_companies_df
```

## IPO Data

```python
# Read in our IPOS_2008 data
IPOS_2008_df = pd.read_csv("IPOS_2008.csv")

IPOS_2008_df
```

```python
# Retrieve IPO Symbol from IPOS_2008.csv and add to a new dataframe, 2008_IPOS_Tickers_df
IPOS_2008_df.Symbol.unique()
```

```python
# Rename Columns
IPOS_2008_df = IPOS_2008_df.rename(columns={'Symbol': 'ticker', 'Date': 'date'})
IPOS_2008_df
```

```python
IPOS_2008_df = IPOS_2008_df[['ticker', 'Company Name', 'date']]
IPOS_2008_df
```

## Merge Data

```python
ipo_info_df = pd.merge(IPOS_2008_df, us_companies_df, how='inner', on=['ticker'])
```

```python
ipo_info_df
```

```python
# Asumption 80 IPOS are missing data from the US Companies.csv file 
ipo_info_2_df = pd.merge(IPOS_2008_df, us_companies_df, how='left', on=['ticker'])
ipo_info_2_df
```

```python
ipo_info_2_df.isnull().sum(axis = 0)
```

```python
#Drop Column/Rename 
ipo_info_df = ipo_info_df.drop(["Company Name", "short name", "tag 1", "tag 2", "tag 3", "logo"], axis=1)
ipo_info_df
```

```python
ipo_info_df
```

```python
ipo_info_df = ipo_info_df[['ticker', 'date', 'company name', 'industry', 'sector', 'description', 'exchange', 'market cap', 'website', 'ceo']]
ipo_info_df
```

```python
ipo_info_df.to_csv('ipo_info.csv', sep='\t', encoding='utf-8')
```
